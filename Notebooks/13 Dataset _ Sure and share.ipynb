```python
def get_df(soup):
  arr = []
  for i in soup.find_all('a', {"id": "video-title"}):
    d={}
    print(i)
    d['headline'] = i['title']
    d['url'] = i['href']
    
    arr.append(d)
  return pd.DataFrame(arr)
```
