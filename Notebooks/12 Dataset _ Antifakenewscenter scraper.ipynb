```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
source = ''
for i in json_data['html']:
  source += i

def get_df(soup):
  arr = []
  for i in soup.find_all('div', class_='blog-container'):
    d={}
    d['headline'] = i.find('div', class_='blog-title').text.strip()
    d['url'] = i.find('a').get('href')
    d['img_src'] = i.find('img',class_='attachment-post-thumbnail size-post-thumbnail wp-post-image').get('src')
    try:
      d['tag'] = i.find('div',class_='tag-status').get_text()
    except:
      d['tag'] = None
    try:
      d['topic'] = i.find('div',class_='blog-tag tag-interest').get_text()
    except:
      d['topic'] = None
    arr.append(d)
  return pd.DataFrame(arr)
```

```python
soup = BeautifulSoup(source)
df = get_df(soup)
```

```python
import urllib.request
import urllib.parse

def download_img(tag, df):
  for i in range(len(df['img_src'])):
    try:
      url = df['img_src'][i]
      url = urllib.parse.urlsplit(url)
      url = list(url)
      url[2] = urllib.parse.quote(url[2])
      url = urllib.parse.urlunsplit(url)
      urllib.request.urlretrieve(url, "{}/{:04d}.jpg".format(tag, i))
      if i%10==0: print('Progress: {}/{}'.format(i,len(df['img_src'])))
    except Exception as e:
      print('Fetching image no.{} fail'.format(i))
      print(e)
```

```python
download_img('fake',fake)
download_img('fabricate',fabricate)
download_img('true',true)
```
