```python
tokenizer = AutoTokenizer.from_pretrained("airesearch/wangchanberta-base-att-spm-uncased")
model = AutoModelForSequenceClassification.from_pretrained("Nithiwat/soda-berta")
```

```python
from transformers_interpret import SequenceClassificationExplainer
cls_explainer = SequenceClassificationExplainer(
    model,
    tokenizer)
word_attributions = cls_explainer("พฤติกรรมเด็กเล็กกัดเล็บ ดูดนิ้ว เสี่ยงฟันผิดรูป ฟันซ้อนเก ดังนั้น พ่อแม่ผู้ปกครองควรหมั่นสังเกตและใส่ใจดูแลสุขภาพช่องปากเด็กตั้งแต่มีฟันซี่แรก และพบทันตแพทย์ทุก 6 เดือนเพื่อตรวจและรับคำแนะนำ นายแพทย์ไพโรจน์ สุรัตนวนิช รองอธิบดีกรมการแพทย์ กล่าวถึงการกัดเล็บและดูดนิ้วมือของเด็กว่า เป็นพฤติกรรมที่ผู้ปกครองควรให้ความสำคัญและหมั่นใส่ใจ")
```
